### Task 2.2 

Create a new dataframe with the mean *logbodysize* and *trophicposition* of each genera. Sort this data frame by the largest body size. Print the head of this dataframe.

```python
fishStats_df=pd.read_csv('Bloom_etal_2018_Reduced_Dataset.csv', usecols=['taxa', 'logbodysize', 'trophic_position'])
# reading in the csv with only the columns we want
fishStats_df[['genus','species','subspecies']] = fishStats_df['taxa'].str.split("_",expand=True)
# splitting the taxa column by the underscore character into three columns
fishStats_subdf=fishStats_df[['logbodysize','trophic_position','genus']]
# subsetting the data again - only taking the body size, trophic position, and genus columns
fishMeans_df = fishStats_subdf.groupby('genus',as_index=True).agg(['mean','count'])
# grouping by genus, and getting the mean values for trophic position/body size and count values for the number of fish in each genera in this dataset
fishMeans_df.columns = fishMeans_df.columns.map('_'.join)
# cleaning up our data frame - so there are not column names with sub column names
fishMeans_df.sort_values('logbodysize_mean', ascending=False,inplace=True)
# sorting values so the largest body size is in the top row, smallest is in the bottom row
fishMeans_df
# prints the data frame - answers to the question are below the data frame
```
